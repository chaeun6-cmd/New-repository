```python
import seaborn as sns
import matplotlib.pyplot as plt

# 숫자 컬럼만 선택
num = cars.select_dtypes(include="number")

# 상관계수 계산
corr = num.corr()

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()

#연식이 최신일수록 가격이 비싸다는 것을 확인할 수 있음.
```

```python
#결측치 처리
#cars: title에서 브랜드 추출
cars["brand"] = cars["title"].astype(str).str.split().str[0]
#brand: title 이름 바꾸기
if "title" in brand.columns and "brand" not in brand.columns:
    brand = brand.rename(columns={"title": "brand"})

#표기 통일(대문자로 통일/공백 제거)
cars["brand"] = cars["brand"].astype(str).str.upper().str.strip()
brand["brand"] = brand["brand"].astype(str).str.upper().str.strip()

#merge
df = cars.merge(brand, on="brand", how="left")

print("merged shape:", df.shape)
print("country missing rate:", df["country"].isna().mean() if "country" in df.columns else "no country col")
```
